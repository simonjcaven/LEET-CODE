```python
# Use lists for solutions where index is the column and item is the row
# see L13,14 in FIT1029 plus
# https://www.geeksforgeeks.org/dsa/8-queen-problem/

def getNextPossible(partial,n):
    # given current partial, returns list of rows for NEXT queen
    possible = []
    for i in range(n):
        # can't be a row already in partial
        # can't be a row under diagonal attack from those in partial
        if checkPartial(partial, i) and checkDiagonal(partial, i, n):
            possible.append(i)
    return possible

def checkPartial(partial, target):
    for row in partial:
        if target == row:
            return False
    return True

def checkDiagonal(partial, target, n):
    
    # create board
    board = []
    for i in range(n):
        new_row = [0 for i in range(n)]
        board.append(new_row)

    # populate digonals from partial solution
    for col, row in enumerate(partial):
        
        # add diagonal up left
        i = row
        j = col
        while i >= 0 and j >= 0:
            board[i][j] = 'q'
            i -= 1
            j -= 1

        
        # add diagonal down right
        i = row
        j = col
        while i < n and j < n:
            board[i][j] = 'q'
            i += 1
            j += 1

        # add diagonal up right
        i = row
        j = col
        while i >=0 and j < n:
            board[i][j] = 'q'
            i -= 1
            j += 1

        # add diagonal down left
        i = row
        j = col
        while i < n and j >= 0:
            board[i][j] = 'q'
            i += 1
            j -= 1
        
        
    # index into board to check
    if board[target][len(partial)] == 'q':
        return False
    return True
        
```

```python
# recursive function to generate solutions
def nQueens(partial, n, results):
    if len(partial) == n:
        print(partial)
        results.append(list(partial))
    else:
        possibleRows = getNextPossible(partial, n)
        for row in possibleRows:
            partial.append(row)
            nQueens(partial, n, results)
            
            # if there are no more possible rows
            # returns here(recursive call ends), pops off the last partial
            # then continues the loop by trying the next row in this loop
            # with more recursive functions
            partial.pop()
```

## Using Simulated Annealing

```python
# 1) Generate a random board state with a queen in every row
# 2) Define a Cost Function (number of attacking queen pairs)
# 3) Define a perturb state (select one queen (row) and change its column)
# 4) Implement the Simulated Annealing loop and cooling schedule
# 5) For all solutions run multiple times with storage
```

```python
import random
import numpy as np

def cost_function(board, n):
    # returns the number of conflicting queen pairs
    # checks every pair for column or diagonal clash

    # reurns 0 if a solution, max is n choose 2 which is 28 for n =8
    
    pairs = 0
    for i in range(len(board)):
        for j in range(i+1,len(board)):
            if board[i] == board[j]:
                pairs +=1
            else:
                pos1 = (i, board[i])
                pos2 = (j, board[j])
                if check_as_diagonals(pos1, pos2, n):
                    pairs += 1
    return pairs


def check_as_diagonals(q1, q2, n):
    
    # create board
    board = []
    for i in range(n):
        new_row = [0 for i in range(n)]
        board.append(new_row)

    # add queens
    board[q1[0]][q1[1]] = 'Q'
    board[q2[0]][q2[1]] = 'Q'

    # check pos1 diagonals
    
    # diagonal up left
    i = q1[0] - 1
    j = q1[1] - 1
    while i >= 0 and j >= 0:
        if board[i][j] == 'Q':
            return True
        i -= 1
        j -= 1

    # diagonal down right
    i = q1[0] + 1
    j = q1[1] + 1
    while i < n and j < n:
        if board[i][j] == 'Q':
            return True
        i += 1
        j += 1

    
    # diagonal up right
    i = q1[0] - 1
    j = q1[1] + 1
    while i >=0 and j < n:
        if board[i][j] == 'Q':
            return True
        i -= 1
        j += 1

    # diagonal down left
    i = q1[0] + 1
    j = q1[1] - 1
    while i < n and j >= 0:
        if board[i][j] == 'Q':
            return True
        i += 1
        j -= 1

    return False

def perturb(board, n):

    # select a random row
    rand_row = random.randint(0,n-1)

    # change its column (loop so not same)
    rand_col = random.randint(0,n-1)
    while rand_col == board[rand_row]:
        rand_col = random.randint(0,n-1)

    # update and return 
    board[rand_row] = rand_col

    return board
```

```python
def simulated_annealed_queens(c, no_perturbs, n):

    # generate a random board
    queens = [random.randint(0, n-1) for i in range(n)]

    # simulated annealing loops
    while c > 0:
        for i in range(no_perturbs):
            
            # determine current cost
            current_cost = cost_function(queens, n)

            # a solution
            if current_cost == 0:
                print(queens)
                return

            # perturb a new solution
            new_queens = perturb(queens, n)
            new_cost = cost_function(new_queens, n)

            

            # simulated annealing
            delta = new_cost - current_cost
            if delta < 0: # accept if better
                queens = new_queens
            else:
                # accept with prob of exp(-Delta/c), metroplis acceptance criteria
                r = random.uniform(0,1)
                p = np.exp(-delta/c)
                if r <= p:
                    queens = new_queens

        # tightens acceptance criteria
        c = c*0.89 # reduce c by 89%

```

```python
simulated_annealed_queens(50, 1000, 8)
```
